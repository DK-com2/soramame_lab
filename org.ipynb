```python
import pandas as pd

df = pd.read_excel("玉島org_2406_33202AA0_01.xlsx", header=None)
df
```

```python
NO = df.iloc[4:35, 1:25]
NO2 = df.iloc[45:76, 1:25]
NOX = df.iloc[86:117, 1:25]
SPM = df.iloc[127:158, 1:25]
WD = df.iloc[168:199, 1:25]
WS = df.iloc[209:240, 1:25]
TEMP = df.iloc[250:281, 1:25]
HUM = df.iloc[291:322, 1:25]
```

```python
# データフレームを転置し、リストに値を格納
NO_values = NO.values.flatten().tolist()
NO2_values = NO2.values.flatten().tolist()
NOX_values = NOX.values.flatten().tolist()
SPM_values = SPM.values.flatten().tolist()
WD_values = WD.values.flatten().tolist()
WS_values = WS.values.flatten().tolist()
TEMP_values = TEMP.values.flatten().tolist()
HUM_values = HUM.values.flatten().tolist()

# 新しいデータフレームに各カラムとして追加
data = pd.DataFrame({
    'NO': NO_values,
    'NO2': NO2_values,
    'NOX': NOX_values,
    'SPM': SPM_values,
    'WD': WD_values,
    'WS': WS_values,
    'TEMP': TEMP_values,
    'HUM': HUM_values
})

data.head()
```

```python
data["NO"] = pd.to_numeric(data["NO"], errors='coerce')
data["NO2"] = pd.to_numeric(data["NO2"], errors='coerce')
data["NOX"] = pd.to_numeric(data["NOX"], errors='coerce')
data["SPM"] = pd.to_numeric(data["SPM"], errors='coerce')
data.head()
```

```python
# 指定する年と月
year = df.iloc[0, 2]
month = df.iloc[0, 4]

# 1時間単位の日付と時間の範囲を生成（既存データフレームの行数に合わせる）
datetime_range = pd.date_range(start=f'{year}-{month:02d}-01 01:00',
                               periods=len(data), freq='H')

# datetimeの列を追加
data['date_time'] = datetime_range

data = data[["date_time", "NO", "NO2", "NOX", "SPM", "WD", "WS", "TEMP", "HUM"]]

data
```

```python
# 欠損値を削除
data = data.dropna(subset=["NO", "NO2", "NOX", "SPM", "WD", "WS", "TEMP", "HUM"], how='all')
data
```
